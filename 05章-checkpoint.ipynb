## 0503 用pandas熊貓做基本數據分析

![e](pd.jpg)

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("grades.csv")
```

```python
df.國文.mean()  #可以直接算，不需要先輸給numpy
```

```python
df.describe()
#基本統計數據整理 .describe()
```

```python
df.corr()
#各個學科之間的相關係數 .corr()
```

```python
df.國文.corr(df.數學)
#只有國文跟數學的相關係數
```

```python
df["總級分"] = df[["國文","英文","社會","自然","數學"]].sum(1)
#新增一個總級分的欄位
#想要加的方向是行的方向 : sum(1)，列方向sum(0)
```

```python
df.head()
```

![o](sum.jpg)

```python
df["主科"] = df[["國文","英文","數學"]].sum(1)
#只想看主科的分數
```

```python
df.head()
```

```python
df["主科"] = df.數學*1.5 +  df.英文
```

```python
df.head()
```

![o](df2.jpg)

```python
df.sort_values(by="總級分")
#排序
```

```python
df.sort_values(by="總級分",ascending = False).head(20)
#排序；ascending = False : 從大到小；先看前20位同學
```

![o](sort.jpg)

```python
df.sort_values(by=["主科","總級分"],ascending = False).head(20)
#排序 :比完主科再比總級分
```
